**Simple Blurring**

```python
def simpleBlur (image, kernelSize, imName):
  blurred = cv2.blur(image, (kernelSize,kernelSize)) #10, 10 is kernel size
  plt.imshow(blurred.astype('uint8'))
  print ("SIMPLE BLUR: KernelSize: " + str(kernelSize))
  plt.show()
  imName = imName + "SimpleBlur.png"
  cv2.imwrite(imName, blurred) 
```

**Gauss Blur**

```python
def gauss_blur(img, kernelSize, sigma, imName):
  blur = cv2.GaussianBlur(img, (kernelSize, kernelSize), sigma)
  plt.imshow(blur.astype('uint8'))
  print ("GAUSSIAN BLUR: KernelSize: " + str(kernelSize) + " ,sigma= " + str(sigma))
  plt.show()
  imName = imName + "GaussianBlur.png"
  cv2.imwrite(imName, blur) 
```

**Median Filter**

```python
def MedianFilter (img, kernelSize,imName):
  blur = cv2.medianBlur(img, kernelSize)  
  plt.imshow(blur.astype('uint8'))
  print ("MEDIAN FILTER: KernelSize: " + str(kernelSize))
  plt.show()
  imName = imName + "MedianFilter.png"
  cv2.imwrite(imName, blur) 
```

**Bilateral Filter**

```python
def BilateralFilter (img, kernelSize,imName):
  blur = cv2.medianBlur(img, kernelSize)  
  plt.imshow(blur.astype('uint8'))
  print ("MEDIAN FILTER: KernelSize: " + str(kernelSize))
  plt.show()
  imName = imName + "BilateralFilter.png"
  cv2.imwrite(imName, blur) 
```

**Prewitt edge detector**

```python
def PrewittDetector(img, kernel,imName):
  res = cv2.filter2D(img, -1, kernel)
  plt.imshow(res.astype('uint8'))
  plt.show()
  imName = imName + "PrewittDetector.png"
  cv2.imwrite(imName, res) 
```

**Canny Edge Detector**

```python
def cannyDetector(img,imName):
  res = cv2.Canny(img,100,200)
  plt.imshow(res.astype('uint8'))
  print ("CANNY DETECTOR")
  plt.show()
  imName = imName + "cannyDetector.png"
  cv2.imwrite(imName, res) 
```

**Sobel Detector**

```python
def sobelDetector(img, kernelSize,imName):
  res_x = cv2.Sobel(img ,cv2.CV_8U,1,0,kernelSize) #gradient x
  plt.imshow(res_x.astype('uint8'))
  print ("SOBEL DETECTOR_X")
  plt.show()
  res_y = cv2.Sobel(img, cv2.CV_8U, 0 ,1 , kernelSize) #gradient y
  plt.imshow(res_y.astype('uint8'))
  print ("SOBEL DETECTOR_Y")
  plt.show()
  imName_x = imName + "sobelDetector_x.png"
  cv2.imwrite(imName_x, res_x)
  imName_y = imName + "sobelDetector_y.png"
  cv2.imwrite(imName_y, res_y) 
```

**Laplacian Detector**

```python
def Laplacian(img,imName):
  res = cv2.Laplacian(img,cv2.CV_64F)
  plt.imshow(res.astype('uint8'))
  print ("Laplacian DETECTOR")
  plt.show()
  imName = imName + "Laplacian.png"
  cv2.imwrite(imName, res) 
```

**Unsharp Masking**

```python
def unsharpMasking(img, radius, amount,imName):
  res = unsharp_mask(img, radius, amount)
  fig, axes = plt.subplots(nrows=2, ncols=2,
                         sharex=False, sharey=False, figsize=(10, 10))
  ax = axes.ravel()
  ax[1].imshow(res, cmap=plt.cm.gray)
  print ("UNSHARP MASKING: radius: " + str(radius) +  "amount: " + str(amount))
  imName = imName + "unsharpMasking.png"
  cv2.imwrite(imName, res) 
```
